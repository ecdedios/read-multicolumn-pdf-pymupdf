```python
import fitz
```

```python
# global variables (replace with you own files)
DIGITIZED_FILE = "../data/in/Digitized - Multi-Column Single Page Document.pdf"     # searchable
SCANNED_FILE = "../data/in/Scanned - Multi-Column Single Page Document.pdf"         # not searchable
```

# PyMuPDF - Digitized PDFs

```python
# OCR the PDF using the default 'text' parameter
with fitz.open(DIGITIZED_FILE) as doc:
    for page in doc:
        text = page.get_text()
        print(text)
```

```python
# OCR the PDF using the 'blocks' parameter
with fitz.open(DIGITIZED_FILE) as doc:
    for page in doc:
        text = page.get_text("blocks")
        print(text)
```

```python
type(text)
```

```python
type(text[0])
```

```python
text[0] # the first item in the list
```

```python
text[0][0] # first tuple item of the first item in the list
```

```python
# OCR the PDF using the 'words' parameter
with fitz.open(DIGITIZED_FILE) as doc:
    for page in doc:
        text = page.get_text("words")
        print(text)
```
